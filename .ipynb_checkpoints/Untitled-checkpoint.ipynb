```python
import torch
import torchvision
import torchmetrics
from torchmetrics import Accuracy
from torch import nn, optim
from torch.utils.data import DataLoader, SubsetRandomSampler, Dataset
from torchvision import models
from torchvision import transforms as T
from torchvision.utils import make_grid
from torchvision.datasets import ImageFolder
from torchsummary import summary
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from torchutils import print_reports
device = 'cuda' if torch.cuda.is_available() else 'cpu'
torch.cuda.get_device_properties(device)
```

```python
transform = {
    'train' : T.Compose([
        T.Resize(size=224),
        T.ToTensor()
    ]),
    
    'valid' : T.Compose([
        T.Resize(size=224),
        T.ToTensor()
    ])
}
```

```python
train_data = ImageFolder(root='PlantVillage/', transform=transform['train'])
valid_data = ImageFolder(root='PlantVillage/', transform=transform['valid'])
train_data.class_to_idx
```

```python
total_samples = np.arange(len(train_data))
train_idx, valid_idx = train_test_split(total_samples, test_size=.3, stratify=train_data.targets)
test_idx, valid_idx = train_test_split(valid_idx, test_size=.5, stratify=np.array(train_data.targets)[valid_idx])
len(train_idx), len(valid_idx), len(test_idx)
```

```python
batch_size = 32
train_sampler = SubsetRandomSampler(train_idx)
valid_sampler = SubsetRandomSampler(valid_idx)
test_sampler = SubsetRandomSampler(test_idx)
```

```python
test = ['Train Ratios', 'Valid Ratios', 'Test Ratios']
for i, idx in enumerate([train_idx, valid_idx, test_idx]):
    t = np.array(train_data.targets)[idx]
    t = np.unique(t, return_counts=True)[-1] / len(t)
    print(f"{test[i]} : {t}")
```

```python
data_loaders = {
    'train' : DataLoader(dataset=train_data, sampler=train_sampler, batch_size=batch_size, num_workers=2),
    'valid' : DataLoader(dataset=valid_data, sampler=valid_sampler, batch_size=batch_size, num_workers=2),
    'test' : DataLoader(dataset=valid_data, sampler=test_sampler, batch_size=batch_size, num_workers=2),
}
```

```python
for x, _ in data_loaders['train']:
    plt.figure(figsize=(12,12))
    plt.xticks([]); plt.yticks([])
    plt.title('Train Set')
    plt.imshow(make_grid(x, nrow=4, padding=5, pad_value=1).permute(1,2,0))
    break
    
for x, _ in data_loaders['valid']:
    plt.figure(figsize=(12,12))
    plt.xticks([]); plt.yticks([])
    plt.title('Valid Set')
    plt.imshow(make_grid(x, nrow=4, padding=5, pad_value=1).permute(1,2,0))
    break
    
for x, _ in data_loaders['test']:
    plt.figure(figsize=(12,12))
    plt.xticks([]); plt.yticks([])
    plt.title('Test Set')
    plt.imshow(make_grid(x, nrow=4, padding=5, pad_value=1).permute(1,2,0))
    break
```

```python
model = models.vgg16_bn(pretrained=True)
for param in model.parameters():
    param.requires_grad = False
```

```python
model
```

```python
in_features = model.classifier[6].in_features
model.classifier[6] = nn.Sequential(
    nn.Dropout(p=0.5),
    nn.Linear(in_features=in_features, out_features=5),
    nn.Softmax(dim=1)
)
model = model.to(device)
model
```

```python
lr = 3e-3
weight_decay = 1e-4
loss = nn.CrossEntropyLoss()
optimizer = optim.RMSprop(lr=lr, weight_decay=weight_decay, params=model.parameters())
```

```python
def fit(model, loader, optimizer, loss, epochs=5):
    TL, TA, VL, VA =[], [], [], []
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    train_metric = Accuracy().to(device)
    valid_metric = Accuracy().to(device)
    
    for epoch in tqdm(range(epochs), desc='Epochs', leave=True):
        
        model.train()
        cst = 0
        for x, y in tqdm(loader['train'], desc='Train Step'):
            x = x.to(device)
            y = y.to(device)
            preds = model(x).to(device)
            acc = train_metric(preds, y)
            cost = loss(preds, y)
            cst += cost.item()
            cost.backward()
            optimizer.step()
            optimizer.zero_grad()
        
        cst /= len(loader['train'])
        TL.append(cst)
        TA.append(train_metric.compute().item())
        print(f"TL : {cst}\tTA : {train_metric.compute()}")
        model.eval()
        cst = 0
        for x, y in tqdm(loader['valid'], desc='Valid Step'):
            x = x.to(device)
            y = y.to(device)
            preds = model(x).to(device)
            acc = valid_metric(preds, y)
            cost = loss(preds, y)
            cst += cost.item()
        
        cst /= len(loader['valid'])
        VL.append(cst)
        VA.append(valid_metric.compute().item())
        print(f"VL : {cst}\tVA : {valid_metric.compute()}")
        
    plt.figure(figsize=(20, 7))
    plt.subplot(121)
    plt.plot(TL, '-xr', VL, '-xg')
    plt.title('Cost Curve')
    plt.legend(['Train Loss', 'Valid Loss'])
    plt.subplot(122)
    plt.plot(TA, '-xr', VA, '-xg')
    plt.title('Accuracy Curve')
    plt.legend(['Train Acc', 'Valid Acc'])
    plt.tight_layout()
    plt.grid(True)
        
    return model
```

```python
md = fit(model, data_loaders, optimizer, loss, epochs=5)
```

```python
print_reports(md, data_loaders)
```

```python
def print_reports(model, data_loaders, *, cf_matrix=True):
    DEVICE = 'cuda'
    for key in data_loaders.keys():

        if key == 'train':
            model.train()
            predictions, targets = [], []
            print('--------------------------->> Train Report <<---------------------------')
            for x, y in tqdm(data_loaders[key], desc='Train'):
                x = x.to(DEVICE)
                y = y
                targets += y.tolist()
                predictions += model(x).argmax(dim=1).tolist()
            predictions, targets = np.array(predictions), np.array(targets)
            print(predictions.shape, targets.shape)
            report = classification_report(targets, predictions)
            print(report)
            if cf_matrix:
                cf = confusion_matrix(targets, predictions)
                print('Train Confusion Matrix : ')
                print(cf)
                break

        else:

            model.eval()

            if key == 'valid':
                print('--------------------------->> Validation Report <<---------------------------')
                predictions, targets = [], []
                for x, y in tqdm(data_loaders[key], desc='Valid'):
                    x = x.to(DEVICE)
                    y = y
                    targets += y.tolist()
                    predictions += model(x).argmax(dim=1).tolist()
                predictions, targets = np.array(predictions).ravel(), np.array(targets).ravel()
                report = classification_report(targets, predictions)
                print(report)
                if cf_matrix:
                    cf = confusion_matrix(targets, predictions)
                    print('Validation Confusion Matrix : ')
                    print(cf)

            else:
                print('--------------------------->> Test Report <<---------------------------')
                predictions, targets = [], []
                for x, y in tqdm(data_loaders[key], desc='Test'):
                    x = x.to(DEVICE)
                    y = y
                    targets += y.tolist()
                    predictions += model(x).argmax(dim=1).tolist()
                predictions, targets = np.array(predictions).ravel(), np.array(targets).ravel()
                report = classification_report(targets, predictions)
                print(report)
                if cf_matrix:
                    cf = confusion_matrix(targets, predictions)
                    print('Test Confusion Matrix : ')
                    print(cf)
```
